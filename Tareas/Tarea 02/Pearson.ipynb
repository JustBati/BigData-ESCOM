## Paso 2 - Creacion de clasificador

```python
class PearsonClassifier():
    def __init__ (self, list1, list2):
        self.sample = list1
        self.data = list2
        self.mean_sample = 0    
        self.mean_data = 0
        self.sampleXdata = []
        self.sampleXsample = []
        self.dataXdata = []
        self.sum_sampleXdata = 0
        self.sum_sampleXsample = 0
        self.sum_dataXdata = 0

    def mean(self):
        self.mean_sample = sum(self.sample)/len(self.sample)
        self.mean_data = sum(self.data)/len(self.data)
    
    def FsampleXdata(self): #x*y
        for i in range(len(self.sample)):
            self.sampleXdata.append(self.sample[i]  * self.data[i])
    
    def FsampleXsample(self):#x*x
        for i in range(len(self.sample)):
            self.sampleXsample.append(self.sample[i] * self.sample[i])

    def FdataXdata(self):#y*y
        for i in range(len(self.data)):
            self.dataXdata.append(self.data[i] * self.data[i])

    def Fsum_sampleXdata(self):#sum(x*y)
        self.sum_sampleXdata = sum(self.sampleXdata)

    def Fsum_sampleXsample(self):#sum(x*x)
        self.sum_sampleXsample = sum(self.sampleXsample)
    
    def Fsum_dataXdata(self):#sum(y*y)
        self.sum_dataXdata = sum(self.dataXdata)
    
    def run(self):
        self.mean()
        self.FsampleXdata()
        self.FsampleXsample()
        self.FdataXdata()
        self.Fsum_sampleXdata()
        self.Fsum_sampleXsample()
        self.Fsum_dataXdata()
        topEquation = self.sum_sampleXdata - (len(self.sample) * self.mean_sample * self.mean_data) # sum(x*y) - n * x̄ * ȳ
        bottomEquation = ((self.sum_sampleXsample - (len(self.sample) * (self.mean_sample ** 2 ))) * (self.sum_dataXdata - (len(self.data) * (self.mean_data ** 2)))) ** 0.5 # sqrt((sum(x*x) - n * x̄^2) * (sum(y*y) - n * ȳ^2))
        return print (f"The result of the Pearson coeficent is:\t{topEquation / bottomEquation}")
    
```
